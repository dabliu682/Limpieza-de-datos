```python
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np
```

```python
df=pd.read_csv("saber_pro_limpia.csv")
```

```python
df.head()
```

```python
df.columns
```

```python
df.corr(method="pearson")
```

```python
df.corr()
```

```python
names=list(df)
fig = plt.figure()
ax = fig.add_subplot(111)
cax = ax.matshow(df.corr(), vmin=-1, vmax=1, interpolation='none')
fig.colorbar(cax)
ticks = np.arange(0,34,1)
ax.set_xticks(ticks)
ax.set_yticks(ticks)
ax.set_xticklabels(names)
ax.set_yticklabels(names)
fig.set_size_inches(10,10)
plt.show()
```
